```python
from handlebodies import Genus2Handlebody
```

Genus2Handlebody encodes a sutured handlebody $(M,\gamma)$ of genus 2. 

This handlebody is given by gluing two 3-dimensional balls along 3 disks on their boundaries. The boundary of each disk intersects $R_+$ (or $R_-$) in $a$, $b$, and $c$ connected components (intervals), respectively. Also, we have three parameters $r$, $s$, and $t$ that measure how much we twist the corresponding disk before we glue the two balls.

The input of Genus2Handlebody is the list of tuples [$(a,r), (b,s), (c,t)$].

SFH_plot plots the rank of the sutured Floer homology for all possible spinc structures. Since the set of spinc structures is a torsor over $H_1(M) = \mathbb{Z}^2$, we plot a 2-dimensional grid with the corresponding ranks. The grid is triangular with the 3 generators being the curves that cross the 1st and 2nd, 1st and 3rd, and 2nd and 3rd disks. 

```python
M = Genus2Handlebody([(4,1),(4,2),(3,0)])
plt = M.SFH_plot()
```

```python
# So far Genus2Handlebody only works for inputs that give the generic train track in the pair of pants with
# one of the triangular regions on R_+ and one on R_-.

M = Genus2Handlebody([(5,2),(5,2),(5,3)])
plt = M.SFH_plot()
```

```python
# If you don't want to use the precomputed values, you have to use SFH(load=False) first.
# The program will compute it automatically if SFH was not computed for those parameters.

M = Genus2Handlebody([(5,4),(5,3),(5,2)])
M.SFH(load=False)
plt = M.SFH_plot()
```
