```python
from handlebodies import Genus2Handlebody
```

Genus2Handlebody encodes a sutured handlebody (M,g) of genus 2. 

This handlebody is given by gluing two 3-dimensional balls along 3 disks on their boundaries. The boundary of each disk intersects R_+ (or R_-) in a, b, and c connected components (intervals), respectively. Also, we have three parameters r, s, and t that measure how much we twist the corresponding disk before we glue the two balls.

The input of Genus2Handlebody is the list of tuples [(a,r), (b,s), (c,t)].

Example: (SFH of this example is shown below)


SFH_plot plots the rank of the sutured Floer homology for all possible spinc structures. Since the set of spinc structures is a torsor over H_1(M) = ZZ^2, we plot a grid with the corresponding ranks.

```python
M = Genus2Handlebody([(4,1),(4,2),(3,0)])
plt = M.SFH_plot()
```

```python
# So far Genus2Handlebody only works for inputs that give the generic train track in the pair of pants.
# There are also some bugs for very small values (a, b, or c = 1 for example).
M = Genus2Handlebody([(5,4),(5,3),(5,2)])
plt = M.SFH_plot(xlim=(-1,5), ylim=(-1,5))
```
